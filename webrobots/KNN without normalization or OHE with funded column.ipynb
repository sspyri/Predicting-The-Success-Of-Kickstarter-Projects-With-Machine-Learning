```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score

%matplotlib inline

training_accuracy = []
test_accuracy = []

df = pd.read_csv(r'C:\Users\dungeon\Downloads\ml_data\18kprojects\MissingDataSolutions\mean\categoricalDataLabelEncoder2.csv')

X = df.drop(columns=['Id','Creator','Location','Start','End','Start_Timestamp_(UTC)',
            'End_Timestamp_(UTC)','State','Pledged','Backers','Funded_Percentage'])

y = df['State'].values

# Split dataset into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42,stratify=y)

# Range of n_neighbors tested
neighbors_settings = range(1,150)

# Create KNN classifier for range of n_neighbors
for num_neighbors in neighbors_settings: 
     knn = KNeighborsClassifier(n_neighbors = num_neighbors)

# Fit the classifier to the data

     knn.fit(X_train, y_train)
     
     training_accuracy.append(knn.score(X_train, y_train))

     test_accuracy.append(knn.score(X_test, y_test))
     

# Check accuracy of our model on the test data
     print(num_neighbors, knn.score(X_test, y_test))
    
# Data visualization - Evaluatin the accuracy of both the training and the testing sets against
# n_neighbors
plt.plot(neighbors_settings, training_accuracy, label='Accuracy of the Training Test')
plt.plot(neighbors_settings, test_accuracy, label='Accuracy of the Test Set')
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors')
plt.legend()
```
